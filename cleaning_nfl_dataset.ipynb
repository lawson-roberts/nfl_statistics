```python
pd.set_option('display.max_columns', None)

#df = pd.read_csv('data/nfl_dataset_2002-2019week6.csv')
df = pd.read_csv('data/nfl_historical_game_detail.csv')
# df = df.drop(columns='Unnamed: 0')
df.info()
```

## looks to be alot of fractions we will want to convert, lets write a function to clean these all up

```python
def convert_fraction(column_name, new_column_name):
    # new data frame with split value columns
    new = df[column_name].str.split("-", n = 1, expand = True)
  
    # making separate first name column from new data frame
    df["numerator"]= new[0]
    df["numerator"] = df["numerator"].astype(int)
  
    # making separate last name column from new data frame
    df["denominator"]= new[1]
    df["denominator"] = df["denominator"].astype(int)

    df[new_column_name] = df["numerator"]/df["denominator"]

    # Dropping old Name columns
    df.drop(columns =[column_name, 'numerator', 'denominator'], inplace = True)
    return df
```

```python
df = convert_fraction("third_down_away", "third_down_away")
df = convert_fraction("third_down_home", "third_down_home")
df = convert_fraction("fourth_down_away", "fourth_down_away")
df = convert_fraction("fourth_down_home", "fourth_down_home")
df = convert_fraction("comp_att_away", "comp_att_away")
df = convert_fraction("comp_att_home", "comp_att_home")
df = df.fillna(0)
df.head()
```

```python
def convert_att_yards(column_name):
    # new data frame with split value columns
    new = df[column_name].str.split("-", n = 1, expand = True)
  
    # making separate first name column from new data frame
    att_col_str = column_name + "_occur"
    yards_col_str = column_name + "_yards"

    df[att_col_str]= new[0]
    df[att_col_str] = df[att_col_str].astype(int)
  
    # making separate last name column from new data frame
    df[yards_col_str]= new[1]
    df[yards_col_str] = df[yards_col_str].astype(int)

    # Dropping old Name columns
    df.drop(columns =[column_name], inplace = True)
    return df
```

```python
## need to clean a few other variables that are being treated as str's
df = convert_att_yards("sack_yards_lost_away")
df = convert_att_yards("sack_yards_lost_home")
df = convert_att_yards("penalty_away")
df = convert_att_yards("penalty_home")

## change possession into float
df['possession_away'] = df['possession_away'].str.replace(':', '.')
df['possession_home'] = df['possession_home'].str.replace(':', '.')
df['possession_away'] = df['possession_away'].astype(float)
df['possession_home'] = df['possession_home'].astype(float)
df.head()
```

```python
## needing to declare the winning team to create target variable
df['target'] = np.where(df['away_score'] > df['home_score'], 1, 0)
df['winning_team'] = np.where(df['away_score'] > df['home_score'], df['away_team'], df['home_team'])
df.head()
```

```python
game_urls = pd.read_csv('data/game_id_all.csv')
# game_urls = game_urls.drop(columns = 'Unnamed: 0')
game_urls.head()
```

```python
df = df.merge(game_urls, on='id', how='inner')
df.tail()
```

```python
## dropping a few last un-needed cloumns
df_clean = df.drop(columns=['red_zone_att_away', 'red_zone_att_home'])
df_clean.to_csv('data/nfl_historical_clean.csv', index=False)
print(df_clean.groupby('target').size())
df_clean.head()
```
